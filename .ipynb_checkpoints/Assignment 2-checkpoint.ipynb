```python
"""
Diploma Program class: 
Attributes
- name = "Msc. Data Science"
- prog_courses: List of programs x, must be of class Course
- prog_students: List of students enrolled in the program

Main Methods
- Add course (course object): Creates new Course instance (class - course), adds to course list
- Remove course (course object): Deletes course object from list
- Get_Students: returns list of students enrolled in program
- Get_Program_Courses: returns list of courses in the program
- Students_Passed: Returns list of students who fullfill program pass criteria and prints message if why not
- Students_Passed_WD: Returns list of students who fullfill program pass criteria with distinction
Secondary methods (used to control if student passed)
- Curr.Students: Returns dict of students containing individual dict with student courses and assignments status
- Get_Student_status: Returns list containing numbers of assignments passed for each course
"""

class DiplomaProgramme(object):
    def __init__(self):
        self.prog_courses = []
        self.prog_students = []
        self.name = "MSc. Data Science"
    
    def Get_Students(self):
        return self.prog_students

    #outputs list of programs
    def Get_Program_Courses(self):
        return self.prog_courses
    
    #Adds course object to course list in program
    def Add_Course(self, course):
        assert isinstance(course,Course),"This is not an instance of Course"
        spicy_course = ""
        spicy_course = Course(course,self)
        
        print(spicy_course, "added to program", self.name )
        
        pass
        
    #Removes course from course list in program
    def Remove_Course(self, course):
        assert isinstance(course,Course),"This is not an instance of Course"
        
        #Removing course from students enrolled
        for student in course.Get_Students():
            student.courses.remove(course)
        #Removing students from course list
        self.prog_courses.remove(course)
        
        print(course, "Removed from", self.name)
        pass
    
    #Secondary method: Creating dict of students with number of completed assignments for all courses the student is enrolled in
    def Curr_Students(self):
        students = {}
            
        for student in self.prog_students:
            students[student] = self.Get_Student_Status(student)
        return(students)
    
    #Secondary method: Creating list of nr of passed assignments in each course the student is enrolled in
    def Get_Student_Status(self,student):
        assert student isinstance(student,Student), "You did not enter a student value"
        fp = []
        for i in student.Get_Grades():
            fp.append(i.count(True))
        return fp
    
    #Evaluating dictionary of students and their nr of passed assignments
    def Students_Passed(self):
        #Pulling dict of students and passed assignments
        students_grades = self.Curr_Students()
        pass_students = []
        
        #iterating over each student in the dict
        for student in students_grades:
            student_passes = True
            total = sum(students_grades[student])
            #iterating over nr of passed assignments for each course, if less than 3 student does not pass
            for i in students_grades[student]: 
                if i < 3:
                    student_passes = False
                    break
                else:
                    continue
            
            #evaluating conditions for student fulfilling graduation requirements, if they do, add to list of passed students
            if student_passes == True and total >= 12:
                pass_students.append(student)
                print(student, "passed with ", total, "completed assignments")
            elif student_passes == True: 
                print(student, "did not pass as student did not finish 4 courses")
            else:
                print(student, "did not pass because the student haven't completed 3 assignments in atleast 1 course")
            
        #if sum(students_grades[student]) >= 17 
        print("these students pass: " + str(pass_students))
        return pass_students
    
    #Using list of passed students from previous function and dict of passed assignments from Curr_Students to evaluate completed assignments
    #and determine if student passed with distinction
    def Students_Passed_WD(self):
        student_grades = self.Curr_Students()
        pass_students_wd = []
        
        #Iterating over list of passed students, control if completed assignments > 17
        for student in self.Students_Passed():
            if sum(student_grades[student]) >= 17:
                pass_students_wd.append(student)
                
        #iterating over students passed with distinction from list generated above to display their name and score
        for wd_student in pass_students_wd:
            print( wd_student, "passes with distinction, completed assignments:", sum(student_grades[student]) )
        return pass_students_wd
                
        

        
    


```

```python
"""
Course Class:
ATTRIBUTES
- name: Name of course
- students: students enrolled (initially empty)
- id: Course ID
- Program: Program name, initially set to data_sci
METHODS
- Set_ID: Sets ID of course, "course" + counter starting at 100
- Get_ID: returns course ID str
- Get_Students: returns list of students enrolled in course
- Add_Student(student): Adds student to course, if student not enrolled in program yet, adds student to program prog_students
- Add_All_Students: Adds all students enrolled in the course program to the course
- Remove_Student: Removes student from course
- Remove_All_Students: Removes all students from the course
- Pass_Assignment(assignment nr, student): Changes element in assignment list for student to True
- Pass_All(student): Passes all assignments of the student for the course 
- Fail_Assignment(assignment nr, student): Changes element in assignment list for student to False
- Fail_All(student): Fails all assignments of the student for the course
"""

class Course(object):
    
    #Defining class variables
    assignments_list = ["NaN","NaN","NaN","NaN","NaN"]
    #course ID starting at 100
    c_id_trace = 100
    
    def __init__(self, name, program):
        #Asserting name is str
        assert isinstance(name,str) and len(name) > 0, "Course name not valid"
        #Asserting student is type Student
        self.name = name
        self.students = []
        self.id = self.Set_Id()
        self.program = program
        
        program.prog_courses.append(self)
        
    def __repr__(self):
        return self.name
    
    #Adding course identifier
    def Set_Id(self):
        c_id = "course" + str(self.c_id_trace)
        self.c_id_trace += 1
        return c_id
    
    #returns list of students enrolled
    def Get_Students(self):
        return self.students

    #Adds student to course and assigns 5 assignments specific to the course. If student is not enrolled in program, the student will be enrolled
    def Add_Student(self, student):
        assert isinstance(student,Student), "Student is not a Student object"
        assert student not in self.students, "student is already enrolled"
        
        self.students.append(student)
        student.Set_Assignments(self)
        
        #Testing if student is enrolled in program, enrolls if not
        if student not in self.program.Get_Students():
            self.program.prog_students.append(student)
        
        print(student, "added to", self.name)
        pass
    
    #Adds all students in program to the course
    def Add_All_Students(self):
        for student in self.program.prog_students:
            if student not in self.students:
                self.Add_Student(student)
            else:
                continue
        print("Students:", self.students, "Added to", self.name)
        pass
    
    #Removes student from the course, removes assignments
    def Remove_Student(self, student):
        assert isinstance(student,Student), "Student is not a Student object"
        
        self.students.remove(student)
        student.Remove_Assignments(self)
        
        print(student, "removed from", self.name)
        pass
    
    #Removes all students from the course
    def Remove_All_Students(self):
        for student in self.students[::-1]:
            self.Remove_Student(student)
    
    #Passes assignment (specified by nr) of specified student
    def Pass_Assignment(self, assignment,student):
        assert isinstance(assignment, int) and 0 < assignment =< 5 , "assignment nr needs to be an int between 1 and 5"
        assert isinstance(student,Student), "Student is not a Student object"
        student.assignments[self][assignment-1] = True
        print (student, "passed assignment", assignment)
        pass
    
    #
    def Pass_All(self, student):
        assert isinstance(student,Student), "Student is not a Student object"
        
        student.assignments[self] = [True,True,True,True,True]
        print (student, "passed all assignments")
        pass
        
    def Fail_Assignment(self,assignment, student):
        assert isinstance(assignment, int) and 0 < assignment =< 5 , "assignment nr needs to be an int between 1 and 5"
        assert isinstance(student,Student), "Student is not a Student object"
        
        student.assignments[self][assignment-1] = False
        print(student, "failed assignment", assignment)
        pass
    
    def Fail_All(self,student):
        student.assignments[self] = [False,False,False,False,False]
        print(student, "failed all assignments")
        pass
    
```

```python
"""
Student: 
Student information:
- Name: Must be string and contain atleast one symbol
- Student ID: Automaticly added, format "stu" + number starting from 100
- Must be valid program
METHODS
- __str__: specifies students name, ID, courses and program 
- Get_Student_Courses(): retrievs list of courses student is enrolled in
- Assignments_Status(course): retrieves DataFrame with status of assignments a course
- Assignments_Status_All(): returns dictionary of passed, failed and non-taken courses for all courses student is enrolled in
- Total

"""


class Student(object):
    id_trace = 100
    
    def __init__(self, fname, lname, program = data_sci):
        
        #Checking name and birthday is in correct format
        assert isinstance(fname,str) and len(fname) > 0, "Student name not valid"
        assert isinstance(lname,str) and len(lname) > 0, "Student name not valid"
        assert type(program).__name__ == "DiplomaProgramme", "This program name is not valid"
        
        self.fname = fname
        self.lname = lname
        self.name = str(fname + " " + lname)
        self.program = program
        
        self.id = self.Set_ID()
        self.assignments = {}
        
        self.program.prog_students.append(self)
        
        #Assigning student ID
        
    def Set_ID(self):
        s_id = "student" + str(self.id_trace)
        self.id_trace += 1 
        return s_id
    
    def Get_ID(self):
        print(self.name, "ID:", self.id)
    
    def Set_Assignments(self, course):
        assert course not in self.assignments, "Student already enrolled in course"
        self.assignments[course] = course.assignments_list
        pass
    
    def Remove_Assignments(self,course):
        assert course in self.assignments, "Student is not enrolled in course"
        del self.assignments[course]
        pass
    
    def Get_Courses(self):
        courses = []
        for course in self.assignments:
            courses.append(course)
        print(self.name, "is enrolled in courses:", courses)
        return courses
    
    def Get_Grades(self):
        grades = []
        for course in self.assignments:
            grades.append(self.assignments[course])
        return grades
    
    def Show_Grades(self):
        print(self.name, "Grades: key - (True = pass, False = Fail, NaN = Not graded)")
        for course in self.assignments:
            print(course,"assignments:", self.assignments[course])
        pass
    
    def Get_Info(self):
        print("Student name: " + self.name + ", id: " + self.id + ", enrolled in " + str(self.Get_Courses()) + " with grades " + str(self.Get_Grades()) )
        
    def __repr__(self):
        return self.name
        
    def __str__(self):
        return self.name
        
    
#    def get_info(self):
#        return "Student: " + self.name + ", ID:" + str(self.id) + ", is enrolled in: " \
#        + str(self.courses) + ", in programme " + self.program.name + ". He has passed "
    
```
