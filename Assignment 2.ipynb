```python
class Course(object):
    
    #Defining class variables
    assignments_list = [0,0,0,0,0]
    c_id_trace = 100
    
    def __init__(self, name, program = data_sci):
        #Asserting name is str
        assert isinstance(name,str) and len(name) > 0, "Course name not valid"
        #Asserting student is type Student
        self.name = name
        self.students = []
        self.id = self.Set_Id()
        self.program = program
        
        program.prog_courses.append(self)
        
    def __repr__(self):
        return self.name
    
    #Adding course identifier
    def Set_Id(self):
        c_id = "course" + str(self.c_id_trace)
        self.c_id_trace += 1
        return c_id
    
    #returns list of students enrolled
    def Get_Students(self):
        return self.students

    
    def Add_Student(self, student):
        assert student not in self.students, "student is already enrolled"
        
        self.students.append(student),
        student.Set_Assignments(self)
        
        print(student, "added to", self.name)
        pass
    
    def Add_All_Students(self):
        for student in self.program.prog_students:
            if student not in self.students:
                self.Add_Student(student)
            else:
                continue
        print("Students:", self.students, "Added to", self.name)
        pass
    
    def Remove_Student(self, student):
        self.students.remove(student)
        student.Remove_Assignments(self)
        
        print(student, "removed from", self.name)
        pass
    
    def Remove_All_Students(self):
        for student in self.students[::-1]:
            self.Remove_Student(student)
            
    def Pass_Assignment(self, assignment,student):
        student.assignments[self][assignment-1] = True
        print (student, "passed assignment", assignment)
        pass
    
    def Pass_All(self, student):
        student.assignments[self] = [True,True,True,True,True]
        print (student, "passed all assignments")
        pass
        
    def Fail_Assignment(self,assignment, student):
        student.assignments[self][assignment-1] = False
        print(student, "failed assignment", assignment)
        pass
    
    def Fail_All(self,student):
        student.assignments[self] = [False,False,False,False,False]
        print(student, "failed all assignments")
        pass
    
```

```python
"""
Student: 
Student information:
- Name: Must be string and contain atleast one symbol
- Student ID: Automaticly added, format "stu" + number starting from 100
- Must be valid program
METHODS
- __str__: specifies students name, ID, courses and program 
- Get_Student_Courses(): retrievs list of courses student is enrolled in
- Assignments_Status(course): retrieves DataFrame with status of assignments a course
- Assignments_Status_All(): returns dictionary of passed, failed and non-taken courses for all courses student is enrolled in
- Total

"""


class Student(object):
    id_trace = 100
    
    def __init__(self, fname, lname, program = data_sci):
        
        #Checking name and birthday is in correct format
        assert isinstance(fname,str) and len(fname) > 0, "Student name not valid"
        assert isinstance(lname,str) and len(lname) > 0, "Student name not valid"
        assert type(program).__name__ == "DiplomaProgramme", "This program name is not valid"
        
        self.fname = fname
        self.lname = lname
        self.name = str(fname + " " + lname)
        self.program = program
        
        self.id = self.Set_ID()
        self.assignments = {}
        
        self.program.prog_students.append(self)
        
        #Assigning student ID
        
    def Set_ID(self):
        s_id = "student" + str(self.id_trace)
        self.id_trace += 1 
        return s_id
    
    def Set_Assignments(self, course):
        assert course not in self.assignments, "Student already enrolled in course"
        self.assignments[course] = course.assignments_list
        pass
    
    def Remove_Assignments(self,course):
        assert course in self.assignments, "Student is not enrolled in course"
        del self.assignments[course]
        pass
    
    def Get_Courses(self):
        courses = []
        for course in self.assignments:
            courses.append(course)
        print(self.name, "is enrolled in courses:", courses)
        pass
    
    def Get_Grades(self):
        grades = []
        for course in self.assignments:
            grades.append(self.assignments[course])
        return grades
    
    def Show_Grades(self):
        print(self.name, "Grades: key - (True = pass, False = Fail, NaN = Not graded)")
        for course in self.assignments:
            print(course,"assignments:", self.assignments[course])
        pass
    
    def __repr__(self):
        return self.name
        
    def __str__(self):
        return self.name
        
    
#    def get_info(self):
#        return "Student: " + self.name + ", ID:" + str(self.id) + ", is enrolled in: " \
#        + str(self.courses) + ", in programme " + self.program.name + ". He has passed "
    
```
